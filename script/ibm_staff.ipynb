### Convert columns with numerical ratings to categorical

```python
num_ratings = ['Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 
               'JobSatisfaction', 'PerformanceRating', 'RelationshipSatisfaction', 
               'StockOptionLevel', 'WorkLifeBalance']

for column in num_ratings:
    df[column] = df[column].astype('category')
```

### Add new columns 'AgeGroup' & 'SalaryBracket'

```python
# Categorise age into age groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 20, 30, 40, 50, 100],
                       labels=['<20', '20-30', '30-40', '40-50', '>50'])

df[['Age', 'AgeGroup']].head()
```

```python
# Define the income brackets
bins = [1000, 3000, 7000, 10000, 15000, 20000]  # Define bin edges
labels = ['Very low', 'Low', 'Medium', 'High', 'Very high']  # Define labels for each bracket

# Create the SalaryBracket column based on MonthlyIncome
df['SalaryBracket'] = pd.cut(df['MonthlyIncome'], bins=bins, labels=labels, right=False, include_lowest=True)

# Display the DataFrame with the new SalaryBracket column
df[['MonthlyIncome', 'SalaryBracket']].head()
```

## Outlier Analysis

```python
# Initialize an empty list to store results
outlier_data = []

# Loop through each numerical column
for col in df_numerical.columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_numerical[col].quantile(0.25)
    Q3 = df_numerical[col].quantile(0.75)
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Find the outliers
    outliers = df_numerical[(df_numerical[col] < lower_bound) | (df_numerical[col] > upper_bound)]
    
    # Calculate the number of outliers
    num_outliers = outliers.shape[0]
    # Calculate the percentage of outliers
    pct_outliers = (num_outliers / df_numerical.shape[0]) * 100
    
    # Append results for the column
    outlier_data.append({
        'Column': col,
        'Number of Outliers': num_outliers,
        'Percentage of Outliers': pct_outliers,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound
    })

# Convert the results to a DataFrame
outlier_df = pd.DataFrame(outlier_data)

# Display the outlier analysis DataFrame
outlier_df[outlier_df['Number of Outliers'] > 0]
```

### Outlier Aanalysis on 'MonthlyIncome'

```python
q1_monthly_income = np.quantile(df['MonthlyIncome'], 0.25)
q3_monthly_income = np.quantile(df['MonthlyIncome'], 0.75)
iqr_monthly_income = q3_monthly_income - q1_monthly_income

lower_monthly_income = q1_monthly_income - (1.5 * iqr_monthly_income)
upper_monthly_income = q3_monthly_income + (1.5 * iqr_monthly_income)
```

```python
df.columns
```

```python
outliers_monthly_income = df[(df['MonthlyIncome'] < lower_monthly_income) | (df['MonthlyIncome'] > upper_monthly_income)]

outliers_monthly_income = outliers_monthly_income[['Age', 'Attrition', 'DailyRate', 'MonthlyIncome',
                         'Department', 'JobRole', 'JobInvolvement']]
outliers_monthly_income.sort_values('MonthlyIncome', ascending=False)
```

### Additional insight on 'MonthlyIncome'

```python
from scipy.stats import norm

income_mean = np.mean(df['MonthlyIncome'])
income_std_dev = np.std(df['MonthlyIncome'])

percent_earning_less_than = round(norm.cdf(10000, income_mean, income_std_dev) * 100, 2)
amount_90_percent_earn_less_than = round(norm.ppf(0.9, income_mean, income_std_dev), 2)

print(f"%Employees earning less than 10000: {percent_earning_less_than}%")
print(f"Amount 90% of employees earning less than: {amount_90_percent_earn_less_than}")
```
