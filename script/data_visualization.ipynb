# Data Visualization in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\User-PC\Desktop\Data Analysis\Python\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

## Line Graphs. Default graph is a line graph

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot()
```

### Plot for each variable or column

```python
df.plot(kind = 'line', subplots = True)
```

### Add Chart title and axes labels

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

## Bar Charts

```python
df.plot(kind = 'bar')
```

### Stacked bar chart

```python
df.plot(kind = 'bar', stacked = True)
```

### Plot for particular column(s)

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

### Horizontal bars

```python
df.plot.barh(stacked = True)
```

## Scatter plots. Must specify x and y axes

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

### Increase points sizes and change color

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'yellow')
```

## Histogram

```python
df.plot.hist()
```

### Specify bins

```python
df.plot.hist(bins = 20)
```

## Boxplot

```python
df.boxplot()
```

## Area Plot

```python
df.plot.area()
```

```python
df.plot.area(figsize = (10, 5))
```

## Piechart. Specify column

```python
df.plot.pie(y = 'Flavor Rating')
```

### Increase size

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
